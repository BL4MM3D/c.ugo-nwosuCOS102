```python
#Adekunle gold delivery project

import tkinter as tk
from tkinter import messagebox

def welcomeMessage(message):
    # Create a new window for the welcome message
    window = tk.Toplevel(root)
    window.title("Order Complete")
    window.geometry("500x200")
    window.configure(bg="lightblue")

    label_1 = tk.Label(window, text=f"Thank you for your purchase dear customer\n")  # Fixed case
    label_1.pack()
    label_2 = tk.Label(window, text="Your delivery will arrive in 5-6 business days\n")  # Fixed case
    label_2.pack()

def submit():
    try:
        # Retrieve input values
        location = location_entry.get()
        weight = int(weight_entry.get())  # Convert weight to integer
    except ValueError:
        messagebox.showerror("Input Error", "Please enter a valid weight.")
        return

    # Validate location and weight
    if location in ["Ibeju-Lekki", "Ibeju Lekki", "ibeju lekki", "ibeju-lekki"] and weight >= 10:
        welcomeMessage("Your Total delivery fee comes down to 5000 naira")
    elif location in ["Ibeju-Lekki", "Ibeju Lekki", "ibeju lekki", "ibeju-lekki"] and weight < 10:
        welcomeMessage("Your Total delivery fee comes down to 3500 naira")
    elif location in ["Epe", "epe"] and weight >= 10:
        welcomeMessage("Your Total delivery fee comes down to 10,000 naira")
    elif location in ["Epe", "epe"] and weight < 10:
        welcomeMessage("Your Total delivery fee comes down to 5,000 naira")
    else:
        messagebox.showerror("Input Error", "Please enter required parameters")

# Create Main window
root = tk.Tk()
root.title("Welcome to Simi services")
root.geometry("500x200")

# Create Weight label and entry
location_label = tk.Label(root, text="Location (Ibeju-Lekki or Epe):")  # Fixed case
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

# Create Weight label and entry
weight_label = tk.Label(root, text="Weight:")  # Fixed case
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

# Create Purchase Button
Purchase_button = tk.Button(root, text="Purchase", command=submit)
Purchase_button.pack()

# Run the main event loop
root.mainloop()

```
